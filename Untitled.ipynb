```python
import itertools



def make_letter_series(n):
    letters = [chr(97 + i) for i in range(26)]
    return itertools.product(*[letters for i in range(n)])

import timeit
words = []
start = timeit.default_timer()
for word in make_letter_series(4):
    words.append(word)
end = timeit.default_timer()
print("{}s".format(end-start))

total_requests = len(words)

nstart = timeit.default_timer()
temp_timer = nstart
count = 0
for word in make_letter_series(4):
    count += 1
    if (count % 1000 == 0) or (timeit.default_timer() - temp_timer >= 30):
        print("{} / {}".format(count, total_requests))
        temp_timer = timeit.default_timer()
    query_text = "".join(word)
    results = sp.search(q=query_text, type="artist", limit=50, offset=0)
    total = results['artists']['total']
    if total >= 9000:
        print(query_text, total)
nend = timeit.default_timer()
print("Timer: ", nend - nstart)

```
